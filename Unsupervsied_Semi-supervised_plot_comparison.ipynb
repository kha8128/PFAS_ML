```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
import seaborn as sns
from sklearn.datasets import make_classification, make_regression
```

```python
# visualisation imports
import matplotlib.pyplot as plt
np.random.seed(42)
```

```python
sns.set(rc={'figure.figsize': (10, 10)})
sns.set(font_scale=1.5)
sns.set_style('whitegrid')
```

```python
tsne_df = pd.read_csv("umap_df.csv")

fig, ax = plt.subplots()
ax = sns.scatterplot(data=tsne_df.query("PCBA883 == 0"),x="X",y="Y", s=50, color='lightblue')
ax = sns.scatterplot(data=tsne_df.query("PCBA883 == 1"),x="X",y="Y", s=50, color ='red')

plt.title("PCBA-883 Unsupervised")
plt.savefig('883.png')
plt.show()
```

```python
metric_df = pd.read_csv("metric_df.csv")

fig, ax = plt.subplots()
ax = sns.scatterplot(data=metric_df.query("PCBA883 == 0"),x="X",y="Y", s=50, color='lightblue')
ax = sns.scatterplot(data=metric_df.query("PCBA883 == 1"),x="X",y="Y", s=50, color='red')

plt.title("PCBA-883 Semi-supervised")
plt.savefig('883.png')
plt.show()
```
